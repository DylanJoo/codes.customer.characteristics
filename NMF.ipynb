NMF

```python
from sklearn.decomposition import NMF
K = 10
nmf = NMF(n_components= K, beta_loss= 'frobenius', init='random', l1_ratio = 1) #nndsvd更適用於sparse factorization
W = nmf.fit_transform(df_2020)
H = nmf.components_    
```

```python
df_H = pd.DataFrame(H, columns=df_2020.columns)
df_H
```

```python
for i in range(K):
    display(pd.DataFrame(pd.concat([df_H.loc[i, :].T.nlargest(5), df_H[df_H>0].loc[i, :].T.nsmallest(5)])))
```

```python
df_2020= df_2020.values
loss= df_2020- np.dot(W, H)
print(np.sum(loss))
```

```python
#sample data
df_s= df_s[df_s.index.isin(id_s)]
```
